### Function to remove special character,number,link and html tag from the text

```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    text=re.sub('\W*\d\W','',text)
    return text
```

```python
## applies the wordopt function to the text column of the data.

data['text']=data['text'].apply(wordopt)
```

###  preparing the features (x) and the target variable (y)

```python
x=data['text']
y=data['class']
```

```python
## splits the dataset into training and testing sets.

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

### convert the text data into numerical representations. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

### Create functions for classify news articles as either "Fake News" or "Not A Fake News".

```python
def output_label(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test=pd.DataFrame(testing_news)
    new_def_test["text"]=new_def_test["text"].apply(wordopt)
    new_x_test=new_def_test['text']
    new_xv_test=vectorization.transform(new_x_test)
    pred_LR=LR.predict(new_xv_test)
    pred_DT=DT.predict(new_xv_test)
    pred_GBC=GB.predict(new_xv_test)
    pred_RFC=RF.predict(new_xv_test)
    return print("\n\nLR prediction:{} \n DTpredict:{} \n GBC Prdiction: {} \n RFC Prediction: {}".format(output_label(pred_LR[0]),output_label(pred_DT[0]),output_label(pred_GBC[0]),output_label(pred_RFC[0])))
    
```
